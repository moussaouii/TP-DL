# How to download a data set and display an image ?


Ce notebook vise à montrer comment télécharger un dataset et comment afficher des images de cette dataset.

```python
import numpy
import torchvision.datasets as dset
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
cifar10 = dset.CIFAR10('./data', train=True, transform=None, target_transform=None, download=True)
```

```python
print('Number of samples: ', len(cifar10))
img, target = cifar10[3] # load 4th sample

print(img)
print(target)

print("image size:", img.size)
print("image target:", target)
print("image mode:", img.mode)
```

```python
fig=plt.figure(figsize=(8, 8))
fig.add_subplot(1, 2, 2)
plt.imshow(img, interpolation='nearest')
fig.add_subplot(1, 2, 1)
plt.imshow(img, interpolation='bicubic')
```
